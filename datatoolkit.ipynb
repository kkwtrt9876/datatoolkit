### 1. Demonstrate three different methods for creating identical 2D arrays in NumPy.
### Provide the code for each method and the final output after each method.

```python
import numpy as np

# Method 1: Using np.zeros to create a 2D array of zeros
array_1 = np.zeros((3, 3))
print("Array created with np.zeros:\n", array_1)

# Method 2: Using np.full to create a 2D array filled with a specific value
array_2 = np.full((3, 3), 5)
print("\nArray created with np.full:\n", array_2)

# Method 3: Using np.ones to create a 2D array of ones
array_3 = np.ones((3, 3))
print("\nArray created with np.ones:\n", array_3)
```

### 3. Explain the following terms:
### 1. The difference between np.array, np.asarray, and np.asanyarray

#### - `np.array`: Always creates a new array object, even if the input is already an array.
#### - `np.asarray`: Converts input to an array if it isn't one, but doesn't copy the array if it already is one.
#### - `np.asanyarray`: Similar to np.asarray, but allows subclasses of numpy arrays (like matrices) to remain unchanged.

### 2. The difference between deep copy and shallow copy


#### - Deep copy: Creates a new copy of the object and all objects it references, independent of the original.
#### - Shallow copy: Creates a new object, but references the same inner objects (modifying those inner objects affects both the copy and the original).

### 5. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array, perform the following operations:
#### a) Extract all even integers from the array.
#### b) Extract all odd integers from the array.


```python


array_6 = np.random.randint(1, 10, (5, 6))
even_integers = array_6[array_6 % 2 == 0]
odd_integers = array_6[array_6 % 2 != 0]
print("\nOriginal array:\n", array_6)
print("\nEven integers:\n", even_integers)
print("\nOdd integers:\n", odd_integers)
```

### 6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:
#### a) Find the indices of the maximum values along each depth level (third axis).
#### b) Perform element-wise multiplication of the array with itself.


```python
array_7 = np.random.randint(1, 10, (3, 3, 3))

indices_max_depth = np.argmax(array_7, axis=2)

elementwise_multiplication = np.multiply(array_7, array_7)

print("\n3D array:\n", array_7)
print("\nIndices of maximum values along the third axis:\n", indices_max_depth)
print("\nElement-wise multiplication of the array with itself:\n", elementwise_multiplication)
```

### 13. Create two NumPy arrays x and y, and perform scatter plot operations


```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 6))

# a) Scatter plot
plt.scatter(x, y, color='red', marker='o', label='Random points')

# b) Horizontal line at y = 0.5
plt.axhline(0.5, linestyle='--', color='blue', label='y = 0.5')

# c) Vertical line at x = 0.5
plt.axvline(0.5, linestyle=':', color='green', label='x = 0.5')

# d) Label the x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# e) Set the title
plt.title('Advanced Scatter Plot of Random Values')

# f) Display a legend
plt.legend()
plt.show()
```

### 14. Create a time-series dataset and plot with different y-axes


```python
date_rng = pd.date_range(start='2023-01-01', periods=100, freq='D')
temp_data = pd.DataFrame({
    'Date': date_rng,
    'Temperature': np.random.randint(20, 35, size=(100)),
    'Humidity': np.random.randint(30, 80, size=(100))
})

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)', color='tab:red')
ax1.plot(temp_data['Date'], temp_data['Temperature'], color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create second y-axis for humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(temp_data['Date'], temp_data['Humidity'], color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title
plt.title('Temperature and Humidity Over Time')
plt.show()
```

### 15. Plot histogram with PDF overlay


```python
data = np.random.normal(0, 1, 1000)

# a) Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# b) Overlay a PDF plot
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)
plt.plot(x, p, 'k', linewidth=2)

# c) Label x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title
plt.title('Histogram with PDF Overlay')
plt.show()
```

### 16. Seaborn scatter plot with color based on quadrants


```python
x = np.random.randn(100)
y = np.random.randn(100)
quadrants = np.where((x >= 0) & (y >= 0), 'Q1',
                     np.where((x < 0) & (y >= 0), 'Q2',
                              np.where((x < 0) & (y < 0), 'Q3', 'Q4')))

sns.scatterplot(x=x, y=y, hue=quadrants, palette='bright')
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrants')
plt.show()
```

### 17. Bokeh line chart of sine wave


```python
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Sine')

p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

show(p)
```

### 18. Bokeh bar chart of random categorical data


```python
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 50, len(categories))

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

p.vbar(x=categories, top=values, width=0.9, color="navy")

p.add_tools(HoverTool(tooltips=[("Value", "@top")]))

p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```
